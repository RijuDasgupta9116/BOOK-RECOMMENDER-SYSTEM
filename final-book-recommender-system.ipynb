```python
df = pd.read_csv('/kaggle/input/7k-books-with-metadata/books.csv')
```

```python
print(df.shape)
df.head()
```

```python
df.drop(columns=['isbn13','isbn10','subtitle','published_year','average_rating','num_pages','ratings_count'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df['categories'] = df['categories'].apply(lambda x:x.split())
```

```python
df['description'] = df['description'].apply(lambda x:x.split())
```

```python
df['authors'] = df['authors'].apply(lambda x:x.replace(' ',''))
df['authors'] = df['authors'].apply(lambda x:x.split())
```

```python
df['tags'] = df['authors']+df['categories']+df['description']
```

```python
df.head()
```

```python
new = df.drop(columns=['authors','categories','description'],axis=1)
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=7000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Harry Potter and the Goblet of Fire')
```

```python
import pickle
pickle.dump(new,open('books_list.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
